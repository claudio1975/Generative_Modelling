```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

```

```python
df2 = pd.read_csv('/content/swmotorcycle.csv')
```

# D2

```python
df2=df2.drop_duplicates()
```

```python
# add ClaimOcc feature
df_2 = df2.copy()
df_2['ClaimOcc'] = np.where(df_2['ClaimNb'] > 0, 1, 0)
# Feature transformation
df_2['Exposure'] = df_2['Exposure'].clip(upper=1)
df_2['VehAge'] = df_2['VehAge'].clip(upper=20)
```

```python
# Split the dataset into train/test split
train80, test20 = train_test_split(df_2, test_size=0.2, random_state=0)
```

```python
train80.columns
```

```python
train80.dtypes
```

```python
obj_cols = [var for var in train80.columns if train80[var].dtype=='object']
train80[obj_cols].describe().T
```

```python
# Iterate over each categorical column and extract unique classes
unique_classes = {}
for col in obj_cols:
    unique_classes[col] = train80[col].unique()

# Display the unique classes for each categorical feature
for col, classes in unique_classes.items():
    print(f"Unique classes in '{col}': {classes}")
```

```python
num_cols=[var for var in train80.columns if train80[var].dtype!='object']
train80[num_cols].describe().T
```

```python
# Assuming 'category_column' is the column by which you want to select records
category_column ='Area'
# Select one record per class
selected_vehage = train80.groupby(category_column).apply(lambda x: x.sample(1)).reset_index(drop=True)
# Display the selected records
selected_vehage
```

```python
# Assuming 'category_column' is the column by which you want to select records
category_column = 'Gender'
# Select one record per class
selected_vehbody = train80.groupby(category_column).apply(lambda x: x.sample(1)).reset_index(drop=True)
# Display the selected records
selected_vehbody
```

```python
# Assuming 'category_column' is the column by which you want to select records
category_column = 'BonusClass'
# Select one record per class
selected_drivage = train80.groupby(category_column).apply(lambda x: x.sample(1)).reset_index(drop=True)
# Display the selected records
selected_drivage

```

```python
# Assuming 'category_column' is the column by which you want to select records
category_column = 'RiskClass'
# Select one record per class
selected_gender = train80.groupby(category_column).apply(lambda x: x.sample(1)).reset_index(drop=True)
# Display the selected records
selected_gender
```

```python
selected_claimocc=train80[train80['ClaimOcc']==1].head(10)
selected_claimocc
```

```python
selected_claimnb=train80[train80['ClaimNb']>=2].head(10)
selected_claimnb
```

```python
# Concatenate the DataFrames
combined_df = pd.concat([
    selected_vehage,
    selected_vehbody,
    selected_drivage,
    selected_gender,
    selected_claimocc,
    selected_claimnb
], ignore_index=True)  # Use ignore_index=True to reset the index in the concatenated DataFrame

# Display the combined DataFrame
combined_df
```
